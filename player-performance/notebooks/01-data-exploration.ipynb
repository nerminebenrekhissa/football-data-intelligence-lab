```python
df_clean = df.dropna(subset=features).copy()
df_clean.shape
```

```python
# 1) Features are the same as before
features = [
    'Age', '90s', 
    'Gls_90', 'Ast_90', 'G+A_90', 'G-PK_90', 'G+A-PK_90',
    'xG_90', 'xAG_90', 'xG+xAG_90', 
    'npxG_90', 'npxG+xAG_90',
    'PrgC', 'PrgP', 'PrgR'
]

# 2) X and y from the CLEAN dataframe
X = df_clean[features]
y = df_clean['good_perf']

# 3) Train/test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# 4) Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5) Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model = LogisticRegression(max_iter=500)
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

y_prob = model.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

auc = roc_auc_score(y_test, y_prob)

plt.figure(figsize=(7,5))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc:.3f})")
plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np

importance = pd.DataFrame({
    'feature': features,
    'coefficient': model.coef_[0]
})

importance['abs_coef'] = importance['coefficient'].abs()
importance.sort_values('abs_coef', ascending=False, inplace=True)

importance
```

```python
plt.figure(figsize=(8,6))
plt.barh(importance['feature'], importance['abs_coef'])
plt.title("Feature Importance (Logistic Regression Coefficients)")
plt.xlabel("Absolute Coefficient Value")
plt.gca().invert_yaxis()
plt.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(model, X_test_scaled, y_test)
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    random_state=42
)

rf_model.fit(X_train, y_train)  # NOTE: no scaling needed
rf_pred = rf_model.predict(X_test)

print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)
gb_pred = gb_model.predict(X_test)

print(confusion_matrix(y_test, gb_pred))
print(classification_report(y_test, gb_pred))
```

```python
models = {
    "Logistic Regression": model,
    "Random Forest": rf_model,
    "Gradient Boosting": gb_model
}

results = []

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

for name, m in models.items():
    preds = m.predict(X_test if name != "Logistic Regression" else X_test_scaled)

    accuracy = accuracy_score(y_test, preds)
    precision = precision_score(y_test, preds)
    recall = recall_score(y_test, preds)
    f1 = f1_score(y_test, preds)

    results.append([name, accuracy, precision, recall, f1])

import pandas as pd

results_df = pd.DataFrame(results, columns=["Model", "Accuracy", "Precision", "Recall", "F1 Score"])
results_df
```

```python
def predict_player(player_name, df_source, model, features, top_n=5):
    """
    Search for a player by name, show their stats, and predict if they are a high performer.
    player_name: string to search 
    df_source: usually df_clean
    model: trained model (final_model)
    features: list of feature columns used by the model
    top_n: number of matching players to show if there are many
    """
    # 1) Find matches
    matches = df_source[df_source['Player'].str.contains(player_name, case=False, na=False)]
    
    if matches.empty:
        print(f"No players found matching: {player_name}")
        return
    
    # 2) If many players match, just show the first few and pick the first for prediction
    if len(matches) > 1:
        print(f"Found {len(matches)} players matching '{player_name}'. Showing first {top_n}:")
        display(matches[['Player', 'Squad', 'Pos', 'Age', '90s', 'G+A_90', 'xG_90', 'xAG_90']].head(top_n))
        print("\nUsing the FIRST player in the above list for prediction.\n")
    
    player_row = matches.iloc[0]
    
    X_player = player_row[features].values.reshape(1, -1)
    
    pred_class = model.predict(X_player)[0]
    pred_proba = model.predict_proba(X_player)[0, 1]  
    
    print("Player:", player_row['Player'])
    print("Squad:", player_row['Squad'])
    print("Competition:", player_row['Comp'])
    print("Position:", player_row['Pos'])
    print("Age:", player_row['Age'])
    print("90s played:", player_row['90s'])
    print("\nKey performance stats:")
    print(f"G+A per 90: {player_row['G+A_90']:.3f}")
    print(f"xG per 90:  {player_row['xG_90']:.3f}")
    print(f"xAG per 90: {player_row['xAG_90']:.3f}")
    print(f"Progressive carries (PrgC): {player_row['PrgC']}")
    print(f"Progressive passes  (PrgP): {player_row['PrgP']}")
    print(f"Progressive recvs   (PrgR): {player_row['PrgR']}")
    
    label_text = "HIGH PERFORMER" if pred_class == 1 else "normal/low performer"
    print("\nModel prediction:")
    print(f" â†’ Predicted class: {pred_class}  ({label_text})")
    print(f" â†’ Probability of being a high performer: {pred_proba:.3f}")
```
