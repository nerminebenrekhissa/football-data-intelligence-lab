```python
import pandas as pd

df = pd.read_csv("../data/processed/players_full_2425.csv")

df.shape
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
import numpy as np

def map_role(pos):
    if pd.isna(pos):
        return "OTHER"
    pos = str(pos)
    if "GK" in pos:
        return "GK"
    if any(tag in pos for tag in ["DF", "CB", "RB", "LB", "WB", "FB"]):
        return "DEF"
    if any(tag in pos for tag in ["MF", "DM", "CM", "AM", "WM"]):
        return "MID"
    if any(tag in pos for tag in ["FW", "ST", "CF", "LW", "RW"]):
        return "ATT"
    return "OTHER"

df["role_group"] = df["Pos"].apply(map_role)

df["role_group"].value_counts()
```

```python

attack_cols = [
    "Gls_90", "Ast_90", "G+A_90", "G-PK_90", "G+A-PK_90",
    "xG_90", "xAG_90", "xG+xAG_90", "npxG_90", "npxG+xAG_90",
    "PrgC", "PrgP", "PrgR", "KP", "PPA"
]

defense_cols = [
    "Tkl", "TklW", "Blocks", "Int", "Tkl+Int", "Clr", "Err", "Recov"
]

possession_cols = [
    "Touches", "Carries", "Mis", "Dis"
]

gk_cols = [
    "GA", "Saves", "Save%", "CS", "CS%", "PKA", "PKsv"
]

discipline_cols = ["CrdY", "CrdR"]

base_cols = ["90s"]
```

```python

def zscore(series):
    return (series - series.mean()) / (series.std(ddof=0) + 1e-9)

for col_list in [attack_cols, defense_cols, possession_cols, gk_cols, discipline_cols, base_cols]:
    for col in col_list:
        if col in df.columns:
            df[f"z_{col}"] = zscore(df[col].fillna(0))
```

```python
def average_z(df, cols):
    cols_z = [f"z_{c}" for c in cols if f"z_{c}" in df.columns]
    if not cols_z:
        return pd.Series(0, index=df.index)
    return df[cols_z].mean(axis=1)

df["att_score"]   = average_z(df, attack_cols)
df["def_score"]   = average_z(df, defense_cols)
df["pos_score"]   = average_z(df, possession_cols)
df["gk_score"]    = average_z(df, gk_cols)
df["usage_score"] = average_z(df, base_cols)

if "CrdY" in df.columns and "CrdR" in df.columns:
    df["cards_raw"] = df["CrdY"].fillna(0) + 5 * df["CrdR"].fillna(0)
    df["card_score"] = -zscore(df["cards_raw"])
else:
    df["card_score"] = 0
```

```python
def compute_performance(row):
    role = row["role_group"]
    
    if role == "GK":
        return (
            0.70 * row["gk_score"] +
            0.10 * row["card_score"] +
            0.20 * row["usage_score"]
        )
    elif role == "ATT":
        return (
            0.50 * row["att_score"] +
            0.20 * row["def_score"] +
            0.20 * row["pos_score"] +
            0.10 * row["card_score"]
        )
    elif role == "MID":
        return (
            0.35 * row["att_score"] +
            0.35 * row["def_score"] +
            0.20 * row["pos_score"] +
            0.10 * row["card_score"]
        )
    elif role == "DEF":
        return (
            0.20 * row["att_score"] +
            0.50 * row["def_score"] +
            0.20 * row["pos_score"] +
            0.10 * row["card_score"]
        )
    else:
        # Fallback
        return (
            0.40 * row["att_score"] +
            0.30 * row["def_score"] +
            0.20 * row["pos_score"] +
            0.10 * row["card_score"]
        )

df["performance_score"] = df.apply(compute_performance, axis=1)
```

```python
threshold = df["performance_score"].quantile(0.80)

df["good_perf"] = (df["performance_score"] >= threshold).astype(int)

df["good_perf"].value_counts()
```

```python
df.to_csv("../data/processed/players_full_2425_with_score.csv", index=False)
```
